# **Глава 1. Машинное обучение: общие сведения и понятия**

*Вопрос 1: Какие числовые признаки называются дискретными, а какие - непрерывными? Приведите собственные примеры.*

Ответ:
Дискретные числовые признаки - некоторые количественные оценки объекта, которые невозможно измерить, но можно посчитать. Например: студенты в аудитории, кружки на столе, цветы в букете.
Непрерывные числовые признаки - некоторые количественные оценки объекта, которые не могут быть подсчитаны, но их можно измерить. Например: длина, вес, сила тока.

*Вопрос 2: Опишите разницу между обучением с учителем и обучением без учителя.*

Ответ: Для обучения с учителем характерно использование обучающего набора данных (тренировочной выборки). Для каждого экземпляра из набора данных есть пары "входные данные / признаки - ожидаемый ответ". В этом случае задачей является поиск модели или алгоритма, который предсказывает ожидаемые целевые ответы. В то же время постанока задачи при обучении без учителя затруднена, так как изначально имеются только данные, в которых необходимо найти закономерности.

*Вопрос 3: Опишите разницу между задачами классификации и регрессии.*

Ответ: Задача классификации предполагает наличие некоторого пространства признаков, у которых есть заранее известная разметка (раскраска) отдельных объектов. Сама задача при этом заключается в построении такой разделяющей кривой, которая способна предсказывать метку или класс объекта. Важным условием использования задачи классификации является конечное множество возможных ответов. В случае с задачей регрессии, её функциональность заключается в построении кривой, проходящей через все тренировочные данные. Важным условием использования задачи регрессии является бесконечное множество возможных ответов.

*Вопрос 8: Найдите частные производные сложной функции* $\ E = (\hat{y} - y)^2 $, $\ \hat{y} = wx + b $ *по переменным* $\ w $ *и* $\ b $.

Ответ: Частные производные функции $\ E $ по переменным $\ w $ и $\ b $: $\ \frac {dE} {dw} = 2(wx + b - y) x $, $\ \frac {dE} {db} = 2(wx + b - y) $.

*Вопрос 9: Найдите частные производные сложной функции* $\ E = y ln \hat{y} + (1 - y) ln(1 - \hat{y}) $, $\ \hat{y} = \frac{1} {1 + e^{-z}} $ *по переменной* $\ z $.

Ответ: Частная производная функции $\ E $ по $\ z $: $\ \frac {dE} {dz} = y(1 - \hat {y}) - (1 - y) \hat {y} $.
